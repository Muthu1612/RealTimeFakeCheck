```python
import os
import cv2
import torch
# from torch.utils.data import Dataset, DataLoader
# from torchvision import transforms
from pathlib import Path
# import mediapipe as mp
import numpy as np
import ffmpeg
```

```python
PARENT_PATH = Path(os.getenv("PARENT_DIR", "."))  # default fallback to current directory
data_path = PARENT_PATH / "data"
# Extract audio and video streams
input_file = ffmpeg.input(str(data_path / "input_vid1.mp4"))
audio_stream = input_file.audio
video_stream = input_file.video

# Process streams (e.g., write to files)
audio_stream.output("audio.mp3").run()
video_stream.output("video_only.mp4").run()
```
